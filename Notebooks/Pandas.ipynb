```python
left = pd.DataFrame({'key':['K0','K8','K2','K3'],'A':['A0','A1','A2','A3'],'B':['B0','B1','B2','B3']})
right = pd.DataFrame({'key':['K0','K1','K2','K3'],'C':['C1','C2','C3','C4'],'D':['C0','C1','C2','C3']})
```

```python
pd.merge(left,right,how = 'outer')
```

```python
pd.merge(left,right,how = 'outer',on='key')
```

```python
pd.merge(left,right,how = 'outer',on='key')
```
